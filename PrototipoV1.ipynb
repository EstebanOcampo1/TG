```python
import pygame
import sys

# Inicializar Pygame
pygame.init()

# Tamaño de pantalla
screen_width, screen_height = 800, 600
screen = pygame.display.set_mode((screen_width, screen_height))
pygame.display.set_caption("El Detective Barkley - La Desaparición")

# Colores
WHITE = (255, 255, 255)
BLACK = (0, 0, 0)
BLUE = (0, 0, 255)
GREEN = (0, 255, 0)
RED = (255, 0, 0)
GRAY = (200, 200, 200)

# Fuentes
font = pygame.font.SysFont(None, 40)
small_font = pygame.font.SysFont(None, 30)

# Cargar imágenes (asegúrate de que las rutas sean correctas)
barkley_img = pygame.image.load('imgs/Barkley.png')
background_img = pygame.image.load('imgs/background.jfif')

# Escalar las imágenes si es necesario
barkley_img = pygame.transform.scale(barkley_img, (350, 350))
background_img = pygame.transform.scale(background_img, (screen_width, screen_height))

# Preguntas y respuestas
questions = [
    {"question": "What is your name?", "options": ["Barkley", "Lola", "Whiskers"], "answer": "Barkley"},
    {"question": "Who is missing the collar?", "options": ["Lola", "Barkley", "Whiskers"], "answer": "Lola"},
    {"question": "Is the collar expensive?", "options": ["Yes", "No", "Maybe"], "answer": "Yes"},
    {"question": "Are you a detective?", "options": ["Yes", "No"], "answer": "Yes"}
]

# Variables para controlar el juego
current_question = 0
score = 0
message = ""
show_message = False
message_timer = 0
finished = False  # Bandera para mostrar si se ha terminado el juego

# Función para mostrar texto en la pantalla
def draw_text(text, font, color, x, y):
    screen_text = font.render(text, True, color)
    screen.blit(screen_text, [x, y])

# Función para mostrar el banner de mensaje
def draw_message_banner(message, color):
    banner_rect = pygame.Rect(200, 250, 400, 100)  # Posición y tamaño del banner
    pygame.draw.rect(screen, GRAY, banner_rect)    # Dibujar rectángulo gris detrás
    pygame.draw.rect(screen, BLACK, banner_rect, 4)  # Bordes negros
    draw_text(message, font, color, 300, 275)  # Centrar el texto dentro del banner

# Bucle principal del juego
running = True
while running:
    # Dibujar el fondo
    screen.blit(background_img, (0, 0))

    # Dibujar al personaje Barkley
    screen.blit(barkley_img, (400, 250))  # Posición en la pantalla

    # Manejo de eventos
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            pygame.quit()
            sys.exit()

    # Mostrar las preguntas si no se ha terminado el juego
    if not finished and current_question < len(questions):
        question_data = questions[current_question]
        draw_text(f"Q: {question_data['question']}", font, BLACK, 50, 50)

        # Mostrar las opciones de respuesta
        y_offset = 150
        for i, option in enumerate(question_data["options"]):
            draw_text(f"{i + 1}. {option}", small_font, BLUE, 100, y_offset)
            y_offset += 50

        # Verificación de respuestas (simplificado)
        keys = pygame.key.get_pressed()
        selected_answer = None
        if keys[pygame.K_1]:
            selected_answer = question_data["options"][0]
        elif keys[pygame.K_2]:
            selected_answer = question_data["options"][1]
        elif keys[pygame.K_3] and len(question_data["options"]) > 2:
            selected_answer = question_data["options"][2]

        # Comprobar si la respuesta es correcta y mostrar mensaje
        if selected_answer is not None:
            if selected_answer == question_data["answer"]:
                message = "Correct!"
                message_color = GREEN
                score += 1
            else:
                message = "Wrong!"
                message_color = RED

            show_message = True
            message_timer = pygame.time.get_ticks()  # Inicia el temporizador del mensaje
            current_question += 1

    # Mostrar mensaje si se respondió
    if show_message:
        draw_message_banner(message, message_color)

        # Mostrar el mensaje por 2 segundos (2000 milisegundos)
        if pygame.time.get_ticks() - message_timer > 2000:
            show_message = False

    # Si ya se terminaron todas las preguntas, mostrar puntaje final
    if current_question >= len(questions) and not show_message:
        finished = True
        screen.fill(WHITE)
        draw_text("¡Has completado el capítulo 1!", font, BLACK, 150, 250)
        draw_text(f"Puntaje final: {score}/{len(questions)}", font, BLACK, 150, 300)

    # Mostrar puntaje durante el juego
    if not finished:
        draw_text(f"Score: {score}", small_font, BLACK, 50, 500)

    # Actualizar la pantalla
    pygame.display.flip()
    pygame.time.delay(200)

# Cerrar Pygame
pygame.quit()
```

```python

import pygame
import sys

# Inicialización de pygame
pygame.init()

# Configuración de la pantalla
screen_width, screen_height = 800, 600
screen = pygame.display.set_mode((screen_width, screen_height))
pygame.display.set_caption("Aprende inglés con acciones")

# Colores
BLACK = (0, 0, 0)
WHITE = (255, 255, 255)
GREEN = (0, 255, 0)
RED = (255, 0, 0)

# Fuentes
font = pygame.font.Font(None, 50)

# Datos del juego
sentence = {"subject": "LOLA", "verb": None, "action": "CRYING"}
options = ["IS", "ARE", "AM", "BE"]
correct_answer = "IS"

# Personaje y fondo
character_image = pygame.image.load('imgs/girl_crying.png')  # Imagen del personaje
character_image = pygame.transform.scale(character_image, (300, 300))
background_image = pygame.image.load('imgs/background.jfif')  # Imagen del fondo
background_image = pygame.transform.scale(background_image, (screen_width, screen_height))

# Función para dibujar texto
def draw_text(text, font, color, surface, x, y):
    text_obj = font.render(text, True, color)
    text_rect = text_obj.get_rect(center=(x, y))
    surface.blit(text_obj, text_rect)

# Bucle del juego
def game_loop():
    running = True
    selected_option = None
    feedback = None

    while running:
        screen.fill(WHITE)
        screen.blit(background_image, (0, 0))  # Fondo

        # Dibujar el personaje
        screen.blit(character_image, (250, 100))

        # Dibujar la frase incompleta
        verb_display = sentence["verb"] if sentence["verb"] else "___"
        draw_text(f'{sentence["subject"]} {verb_display} {sentence["action"]}', font, BLACK, screen, screen_width // 2, 50)

        # Dibujar las opciones de respuesta
        option_rects = []
        for i, option in enumerate(options):
            rect = pygame.Rect(150 + i * 150, 450, 100, 50)
            option_rects.append(rect)
            pygame.draw.rect(screen, GREEN if selected_option == option else BLACK, rect, 2)
            draw_text(option, font, BLACK, screen, rect.centerx, rect.centery)

        # Mostrar retroalimentación si se selecciona una opción
        if feedback:
            draw_text(feedback, font, RED if feedback == "Try again!" else GREEN, screen, screen_width // 2, 550)

        # Manejar eventos
        for event in pygame.event.get():
            if event.type == pygame.QUIT:
                pygame.quit()
                sys.exit()
            elif event.type == pygame.MOUSEBUTTONDOWN:
                mouse_pos = pygame.mouse.get_pos()
                for i, rect in enumerate(option_rects):
                    if rect.collidepoint(mouse_pos):
                        selected_option = options[i]
                        if selected_option == correct_answer:
                            sentence["verb"] = correct_answer
                            feedback = "Correct!"
                        else:
                            feedback = "Try again!"

        # Actualizar pantalla
        pygame.display.flip()

# Iniciar el juego
game_loop()
```

```python
import pygame
import sys
import random

# Inicialización de pygame
pygame.init()

# Configuración de pantalla
screen_width, screen_height = 800, 600
screen = pygame.display.set_mode((screen_width, screen_height))
pygame.display.set_caption("Empareja los Nombres con las Imágenes")

# Colores
WHITE = (255, 255, 255)
BLACK = (0, 0, 0)
GREEN = (0, 255, 0)
RED = (255, 0, 0)

# Fuentes
font = pygame.font.Font(None, 40)

# Elementos de emparejamiento
names = ["Cat", "Dog", "Car", "Apple"]
images = ["imgs/Cat.png", "imgs/Dog.png", "imgs/Car.png", "imgs/Apple.png"]
pairs = list(zip(names, images))
random.shuffle(pairs)  # Barajar los pares

# Cargar imágenes
loaded_images = []
for _, img in pairs:
    image = pygame.image.load(img)
    image = pygame.transform.scale(image, (100, 100))
    loaded_images.append(image)

# Cargar imagen de fondo
background_image = pygame.image.load('imgs/background.jfif')  # Ruta de tu imagen de fondo
background_image = pygame.transform.scale(background_image, (screen_width, screen_height))

# Posiciones
name_positions = [(100, 100), (300, 100), (500, 100), (700, 100)]  # Posiciones para los nombres
image_positions = [(100, 400), (300, 400), (500, 400), (700, 400)]  # Posiciones para las imágenes

random.shuffle(image_positions)

# Variables de control
selected_name = None
selected_name_pos = None
selected_image = None
selected_image_pos = None
matches = []  # Lista para guardar los emparejamientos correctos

# Función para dibujar texto
def draw_text(text, font, color, surface, x, y):
    text_obj = font.render(text, True, color)
    text_rect = text_obj.get_rect(center=(x, y))
    surface.blit(text_obj, text_rect)

# Bucle principal del juego
def game_loop():
    global selected_name, selected_image, selected_name_pos, selected_image_pos, matches
    running = True

    while running:
        screen.blit(background_image, (0, 0))  # Dibujar la imagen de fondo

        # Mostrar los nombres
        for i, (name, _) in enumerate(pairs):
            color = GREEN if (name, _) in matches else BLACK
            draw_text(name, font, color, screen, name_positions[i][0], name_positions[i][1])

        # Mostrar las imágenes
        for i, img in enumerate(loaded_images):
            color = GREEN if (pairs[i][0], pairs[i][1]) in matches else BLACK
            rect = pygame.Rect(image_positions[i][0], image_positions[i][1], 100, 100)
            pygame.draw.rect(screen, color, rect, 2)  # Borde de la imagen
            screen.blit(img, rect.topleft)

        # Dibujar las líneas que conectan los emparejamientos correctos
        for match in matches:
            name_index = names.index(match[0])
            image_index = images.index(match[1])
            pygame.draw.line(screen, GREEN, name_positions[name_index], image_positions[image_index], 3)

        # Manejar eventos
        for event in pygame.event.get():
            if event.type == pygame.QUIT:
                pygame.quit()
                sys.exit()
            elif event.type == pygame.MOUSEBUTTONDOWN:
                mouse_pos = pygame.mouse.get_pos()

                # Verificar si un nombre fue seleccionado
                for i, (name, _) in enumerate(pairs):
                    name_rect = pygame.Rect(name_positions[i][0] - 50, name_positions[i][1] - 20, 100, 40)
                    if name_rect.collidepoint(mouse_pos):
                        selected_name = name
                        selected_name_pos = name_positions[i]
                        break

                # Verificar si una imagen fue seleccionada
                for i, img in enumerate(loaded_images):
                    image_rect = pygame.Rect(image_positions[i][0], image_positions[i][1], 100, 100)
                    if image_rect.collidepoint(mouse_pos):
                        selected_image = pairs[i][1]
                        selected_image_pos = image_positions[i]
                        break

                # Verificar si el nombre y la imagen coinciden
                if selected_name and selected_image:
                    for name, image in pairs:
                        if selected_name == name and selected_image == image:
                            matches.append((name, image))  # Emparejamiento correcto
                            selected_name, selected_image = None, None
                            selected_name_pos, selected_image_pos = None, None
                            break

        # Dibujar la línea entre el nombre y la imagen seleccionada, si ambos han sido seleccionados
        if selected_name_pos and selected_image_pos:
            pygame.draw.line(screen, RED, selected_name_pos, selected_image_pos, 3)

        # Actualizar pantalla
        pygame.display.flip()

# Iniciar el juego
game_loop()
```
