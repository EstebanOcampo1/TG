```python
import pygame
import sys

def quinta_parte():
    # Inicializa Pygame
    pygame.init()
    
    # Configuración de la pantalla
    screen_width = 800
    screen_height = 600
    screen = pygame.display.set_mode((screen_width, screen_height))
    pygame.display.set_caption('Novela Visual en Pygame - Parte 5')
    
    # Colores
    WHITE = (255, 255, 255)
    BLACK = (0, 0, 0)
    LIGHT_BLUE = (173, 216, 230)
    DARK_BLUE = (0, 102, 204)
    
    # Carga de imágenes
    background_image = pygame.image.load('imgs/CityAlley.jpg')  # Imagen de fondo
    background_image = pygame.transform.scale(background_image, (screen_width, screen_height))
    
    barkley_image = pygame.image.load('imgs/Barkley-removebg.png')  # Imagen de Barkley
    barkley_image = pygame.transform.scale(barkley_image, (300, 300))

    lolaSinCollar_image = pygame.image.load('imgs/Lola-sincollar.png')  # Imagen de Lola sin collar
    lolaSinCollar_image = pygame.transform.scale(lolaSinCollar_image, (300, 300))
    
    lolaConCollar_image = pygame.image.load('imgs/Lola-removebg.png')  # Imagen de Lola con collar
    lolaConCollar_image = pygame.transform.scale(lolaConCollar_image, (300, 300))

    crow_image = pygame.image.load('imgs/cuervo-removebg.png')  # Imagen del Cuervo Sombra
    crow_image = pygame.transform.scale(crow_image, (300, 300))

    arrow_image = pygame.image.load('imgs/right-arrow.png')  # Imagen de la flecha
    arrow_image = pygame.transform.scale(arrow_image, (50, 50))  # Escala la flecha a un tamaño adecuado

    flashcard_image = pygame.image.load('imgs/6.png')
    flashcard_image = pygame.transform.scale(flashcard_image, (screen_width-100, 400)) 


    # Fuentes para texto
    font = pygame.font.Font(None, 40)
    
    # Diálogos del juego
    dialogues_english = [
        "",
        "We've finally found you, Shadow Crow!",
        "I don't know what you're talking about...",
        "You took my necklace! Please give it back!",
        "Fine! I confess. Here is your necklace.",
        "Justice is served. Let's go home, Lola.",
        "Thank you, Barkley! I'm so happy!"
    ]
    
    dialogues_spanish = [
        "",
        "¡Finalmente te encontramos, Cuervo Sombra!",
        "No sé de qué están hablando...",
        "¡Tú tomaste mi collar! ¡Por favor devuélvemelo!",
        "¡Está bien! Confieso. Aquí está tu collar.",
        "La justicia ha sido servida. Vámonos a casa, Lola.",
        "¡Gracias, Barkley! ¡Estoy tan feliz!"
    ]
    
    # Variables del juego
    current_dialogue = 0
    show_spanish = False
    clock = pygame.time.Clock()

    def draw_text(text, font, color, x, y, max_width):
        """Función para dibujar texto en la pantalla y ajustarlo si es demasiado largo."""
        words = text.split(' ')  # Separa el texto en palabras
        lines = []
        current_line = ""
        
        for word in words:
            # Comprueba si la línea con la palabra actual cabe en el ancho máximo
            test_line = current_line + word + " "
            if font.size(test_line)[0] <= max_width:
                current_line = test_line
            else:
                # Si no cabe, añade la línea actual a las líneas y empieza una nueva
                lines.append(current_line)
                current_line = word + " "
        
        # Añade la última línea
        if current_line:
            lines.append(current_line)
        
        # Dibuja cada línea en la pantalla
        for i, line in enumerate(lines):
            text_surface = font.render(line, True, color)
            screen.blit(text_surface, (x, y + i * 40))  # 40 píxeles entre líneas para evitar solapamientos
    
    def draw_language_button():
        """Dibuja el botón de cambio de idioma en la esquina superior derecha."""
        button_rect = pygame.Rect(screen_width - 120, 20, 100, 40)  # Botón en la esquina superior derecha
        pygame.draw.rect(screen, LIGHT_BLUE, button_rect)  # Fondo del botón
        pygame.draw.rect(screen, DARK_BLUE, button_rect, 2)  # Borde del botón
        draw_text("Ayuda", font, BLACK, button_rect.x + 10, button_rect.y + 10, 100)  # Texto del botón
        return button_rect

    def draw_back_button():
        """Dibuja el botón de retroceso en la esquina superior izquierda."""
        button_rect = pygame.Rect(20, 20, 100, 40)  # Botón en la esquina superior izquierda
        pygame.draw.rect(screen, LIGHT_BLUE, button_rect)  # Fondo del botón
        pygame.draw.rect(screen, DARK_BLUE, button_rect, 2)  # Borde del botón
        draw_text("Atrás", font, BLACK, button_rect.x + 10, button_rect.y + 10, 100)  # Texto del botón
        return button_rect
    
    def draw_arrow_button():
        """Dibuja el botón de la flecha para avanzar a la siguiente parte."""
        if current_dialogue == len(dialogues_english) - 1:  # Si estamos en el último diálogo
            arrow_rect = pygame.Rect(screen_width - 100, screen_height - 100, 50, 50)  # Posición de la flecha
            screen.blit(arrow_image, arrow_rect)  # Dibuja la imagen de la flecha
            return arrow_rect  # Devuelve el rectángulo para la detección de colisiones
        return None  # Si no es el último diálogo, no hay botón

    def game_loop():
        nonlocal current_dialogue, show_spanish
    
        running = True
        language_button = draw_language_button()
        back_button = draw_back_button()
    
        while running:
            for event in pygame.event.get():
                if event.type == pygame.QUIT:
                    pygame.quit()
                    sys.exit()
    
                # Detectar clics
                if event.type == pygame.MOUSEBUTTONDOWN:
                    mouse_pos = pygame.mouse.get_pos()
                    if language_button.collidepoint(mouse_pos):
                        show_spanish = not show_spanish  # Alternar el idioma
                    elif back_button.collidepoint(mouse_pos):
                        if current_dialogue > 0:
                            current_dialogue -= 1
                    elif arrow_button and arrow_button.collidepoint(mouse_pos):  # Si se hace clic en la flecha
                        running = False
                        print("Última parte")
                          # Detener el loop de la novela visual
                    else:
                        # Avanza al siguiente diálogo si no es el botón
                        if current_dialogue < len(dialogues_english) - 1:
                            current_dialogue += 1
    
            # Dibuja el fondo
            screen.blit(background_image, (0, 0))
    
            # Dibuja los personajes según el diálogo actual
            if current_dialogue == 0:
                screen.blit(flashcard_image, (60, 65))
            elif current_dialogue == 1:  # Primer diálogo, Barkley habla
                screen.blit(barkley_image, (50, 150))  # Barkley
            elif current_dialogue == 2:  # Segundo diálogo, los tres personajes
                screen.blit(barkley_image, (50, 150))  # Barkley
                screen.blit(lolaSinCollar_image, (300, 150))  # Lola sin collar
                screen.blit(crow_image, (550, 150))  # Cuervo Sombra
            elif current_dialogue == 3 or current_dialogue == 3:  # Tercer y cuarto diálogo, Lola y Cuervo
                screen.blit(lolaSinCollar_image, (50, 150))  # Lola sin collar
                screen.blit(crow_image, (550, 150))  # Cuervo Sombra
            elif current_dialogue == 5 or current_dialogue == 5:  # Últimos diálogos, Lola con collar y Barkley
                screen.blit(lolaConCollar_image, (50, 150))  # Lola con collar
                screen.blit(barkley_image, (550, 150))  # Barkley
    
            # Dibuja el cuadro de diálogo
            pygame.draw.rect(screen, BLACK, (50, 465, 720, 130))
            pygame.draw.rect(screen, WHITE, (60, 475, 700, 110))
    
            # Dibuja el nombre del personaje
            if current_dialogue in [0, 4]:
                draw_text("Barkley", font, BLACK, 80, 475, 680)  # Nombre de Barkley
            elif current_dialogue == 2 or current_dialogue == 5:
                draw_text("Lola", font, BLACK, 80, 475, 680)  # Nombre de Lola
            else:
                draw_text("Shadow Crow" if not show_spanish else "Cuervo Sombra", font, BLACK, 80, 475, 680)  # Nombre del Cuervo
    
            # Dibuja el diálogo en el cuadro
            dialogue_text = dialogues_english[current_dialogue] if not show_spanish else dialogues_spanish[current_dialogue]
            draw_text(dialogue_text, font, BLACK, 80, 525, 680)  # Texto del diálogo
    
            # Dibuja los botones de idioma y retroceso
            language_button = draw_language_button()
            back_button = draw_back_button()
    
            # Dibuja el botón de la flecha para avanzar a la siguiente parte
            arrow_button = draw_arrow_button()
    
            # Actualiza la pantalla
            pygame.display.flip()
            clock.tick(30)  # Controla la tasa de fotogramas
    
    # Ejecuta el bucle principal del juego
    game_loop()

# Llamada a la función principal

```

```python
import pygame
import sys
def quiz_game():
    # Inicializar Pygame
    pygame.init()
    
    # Tamaño de pantalla
    screen_width, screen_height = 800, 600
    screen = pygame.display.set_mode((screen_width, screen_height))
    pygame.display.set_caption("El Detective Barkley - Investigación del Collar")
    
    # Colores
    WHITE = (255, 255, 255)
    BLACK = (0, 0, 0)
    BLUE = (0, 0, 255)
    GREEN = (0, 255, 0)
    RED = (255, 0, 0)
    
    # Fuentes
    font = pygame.font.SysFont(None, 40)
    small_font = pygame.font.SysFont(None, 30)
    
    # Cargar imágenes (asegúrate de que las rutas sean correctas)
    barkley_img = pygame.image.load('imgs/Barkley-removebg.png')
    background_img = pygame.image.load('imgs/Lilac solid colour.jpg')  # Fondo de la sala de la casa de Lola
    arrow_img = pygame.image.load('imgs/right-arrow.png')  # Imagen de la flecha
    
    # Escalar las imágenes si es necesario
    barkley_img = pygame.transform.scale(barkley_img, (350, 350))
    background_img = pygame.transform.scale(background_img, (screen_width, screen_height))
    arrow_img = pygame.transform.scale(arrow_img, (100, 100))  # Escalar la flecha
    
    # Preguntas y respuestas adaptadas a la historia
    questions = [
        {"question": "What item is missing from Lola?", "options": ["A necklace", "A shirt", "A hat"], "answer": "A necklace"},
        {"question": "Who gave the collar to Lola?", "options": ["Her grandmother", "Her brother", "Barkley"], "answer": "Her grandmother"},
        {"question": "What is Barkley's job?", "options": ["Detective", "Doctor", "Teacher"], "answer": "Detective"},
        {"question": "Where is Barkley going to investigate?", "options": ["Lola's house", "Barkley's office", "The park"], "answer": "Lola's house"}
    ]
    
    # Variables para controlar el juego
    current_question = 0
    message = ""
    show_message = False
    message_timer = 0
    finished = False  # Bandera para mostrar si se ha terminado el juego
    show_arrow = False  # Mostrar flecha al final
    
    # Función para mostrar texto en la pantalla
    def draw_text(text, font, color, x, y):
        screen_text = font.render(text, True, color)
        screen.blit(screen_text, [x, y])
    
    # Función para obtener la opción que fue seleccionada con un clic
    def get_clicked_option(mouse_pos, y_start, options):
        option_height = 50  # Altura de cada opción
        for i, option in enumerate(options):
            if y_start + i * option_height <= mouse_pos[1] <= y_start + (i + 1) * option_height:
                return option
        return None
    
    # Bucle principal del juego
    running = True
    while running:
        # Dibujar el fondo
        screen.blit(background_img, (0, 0))
    
        # Dibujar al personaje Barkley
        screen.blit(barkley_img, (400, 250))  # Posición en la pantalla
    
        # Manejo de eventos
        for event in pygame.event.get():
            if event.type == pygame.QUIT:
                pygame.quit()
                sys.exit()
    
            # Verificar clics del mouse
            if event.type == pygame.MOUSEBUTTONDOWN:
                mouse_pos = pygame.mouse.get_pos()
    
                # Si ya ha terminado el quiz, verificar si se hace clic en la flecha
                if finished and show_arrow:
                    arrow_rect = arrow_img.get_rect(topleft=(650, 500))
                    if arrow_rect.collidepoint(mouse_pos):
                        running = False
                        quinta_parte()
                         # Puedes cambiar esto según lo que quieras que pase al avanzar
    
                # Verificar si una respuesta fue seleccionada
                if not finished and current_question < len(questions):
                    question_data = questions[current_question]
                    selected_answer = get_clicked_option(mouse_pos, 150, question_data["options"])
                    
                    if selected_answer is not None:
                        # Comprobar si la respuesta es correcta
                        if selected_answer == question_data["answer"]:
                            message = "Correct!"
                            message_color = GREEN
                            current_question += 1
                        else:
                            message = "Wrong!"
                            message_color = RED
    
                        show_message = True
                        message_timer = pygame.time.get_ticks()  # Inicia el temporizador del mensaje
    
        # Mostrar las preguntas si no se ha terminado el juego
        if not finished and current_question < len(questions):
            question_data = questions[current_question]
            draw_text(f"Q: {question_data['question']}", font, BLACK, 50, 50)
    
            # Mostrar las opciones de respuesta
            y_offset = 150
            for i, option in enumerate(question_data["options"]):
                draw_text(f"{i + 1}. {option}", small_font, BLUE, 100, y_offset)
                y_offset += 50
    
        # Mostrar mensaje en la parte inferior
        if show_message:
            draw_text(message, font, message_color, 50, 550)
    
        # Si ya se han respondido todas las preguntas, mostrar la flecha
        if current_question >= len(questions):
            finished = True
            show_arrow = True
            screen.blit(arrow_img, (650, 500))  # Dibujar la flecha en la esquina inferior derecha
    
        # Actualizar la pantalla
        pygame.display.flip()
        pygame.time.delay(200)
    
    # Cerrar Pygame
    pygame.quit()

```

```python
#Primer juego (Segunda escena)
import pygame
import sys
import random

def matching_game():
    # Inicialización de pygame
    pygame.init()

    # Configuración de pantalla
    screen_width, screen_height = 800, 600
    screen = pygame.display.set_mode((screen_width, screen_height))
    pygame.display.set_caption("Empareja las Palabras con las Imágenes")

    # Colores
    WHITE = (255, 255, 255)
    BLACK = (0, 0, 0)
    GREEN = (0, 255, 0)
    RED = (255, 0, 0)
    YELLOW = (255, 255, 0)  # Color para destacar la selección

    # Fuentes
    font = pygame.font.Font(None, 40)

    # Elementos de emparejamiento
    names = ["Dog", "Cat", "Brother", "Grandmother"]  # Nombres relevantes
    images = ["imgs/Lola-removebg.png", "imgs/Whiskers-removebg.png", "imgs/brother.png", "imgs/grandma.png"]  # Asegúrate de tener las imágenes correctas
    pairs = list(zip(names, images))

    # Cargar imágenes
    loaded_images = []
    for img in images:
        image = pygame.image.load(img)
        image = pygame.transform.scale(image, (175, 175))  # Ajustar tamaño de imágenes
        loaded_images.append(image)

    arrow_image = pygame.transform.scale(pygame.image.load('imgs/right-arrow.png'), (50, 50))


    # Cargar imagen de fondo
    background_image = pygame.image.load('imgs/Lilac solid colour.jpg')  # Ruta de tu imagen de fondo
    background_image = pygame.transform.scale(background_image, (screen_width, screen_height))

    # Posiciones
    name_positions = [(100, 100), (300, 100), (500, 100), (700, 100)]  # Posiciones para los nombres
    image_positions = [(25, 350), (225, 350), (425, 350), (625, 350)]  # Posiciones para las imágenes
    random.shuffle(image_positions)  # Barajar las posiciones de las imágenes

    # Variables de control
    selected_name_index = None
    selected_image_index = None
    matches = []  # Lista para guardar los emparejamientos correctos
    disabled = [False] * len(pairs)  # Estado de habilitación para nombres e imágenes

    # Función para dibujar texto
    def draw_text(text, font, color, surface, x, y):
        text_obj = font.render(text, True, color)
        text_rect = text_obj.get_rect(center=(x, y))
        surface.blit(text_obj, text_rect)

    # Bucle principal del juego
    def game_loop():
        nonlocal selected_name_index, selected_image_index, matches
        running = True
        arrow_button_visible = False

        while running:
            screen.blit(background_image, (0, 0))  # Dibujar la imagen de fondo

            # Mostrar los nombres
            for i, (name, _) in enumerate(pairs):
                color = GREEN if i in matches else BLACK
                highlight_color = YELLOW if selected_name_index == i else color  # Cambiar color si está seleccionado
                draw_text(name, font, highlight_color, screen, name_positions[i][0], name_positions[i][1])

            if len(matches) == len(pairs):
                arrow_button_visible = True

            if arrow_button_visible:
                arrow_rect = pygame.Rect(screen_width - 100, screen_height - 100, 50, 50)
                screen.blit(arrow_image, arrow_rect)
                
            # Mostrar las imágenes
            for i, img in enumerate(loaded_images):
                color = GREEN if i in matches else BLACK
                highlight_color = YELLOW if selected_image_index == i else color  # Cambiar color si está seleccionado
                rect = pygame.Rect(image_positions[i][0], image_positions[i][1], 175, 175)
                pygame.draw.rect(screen, highlight_color, rect, 2)  # Borde de la imagen
                screen.blit(img, rect.topleft)

            # Dibujar líneas entre los nombres y las imágenes emparejadas
            for name_idx in matches:
                image_idx = name_idx  # La posición del nombre y la imagen son iguales
                pygame.draw.line(screen, GREEN, name_positions[name_idx], image_positions[image_idx], 3)

            # Dibujar líneas para emparejamientos incorrectos
            if selected_name_index is not None and selected_image_index is not None:
                if selected_name_index != selected_image_index:
                    pygame.draw.line(screen, RED, name_positions[selected_name_index], image_positions[selected_image_index], 3)

            

            # Manejar eventos
            for event in pygame.event.get():
                if event.type == pygame.QUIT:
                    pygame.quit()
                    sys.exit()
                elif event.type == pygame.MOUSEBUTTONDOWN:
                    mouse_pos = pygame.mouse.get_pos()

                    if arrow_button_visible:
                        arrow_rect = pygame.Rect(screen_width - 100, screen_height - 100, 50, 50)
                        if arrow_rect.collidepoint(mouse_pos):
                            # Llama a la segunda parte de la novela visual
                            running = False  # Detener el loop de emparejar
                            segunda_parte()
                            
                    # Verificar si un nombre fue seleccionado
                    for i, (name, _) in enumerate(pairs):
                        if not disabled[i]:  # Solo permitir seleccionar si no está deshabilitado
                            name_rect = pygame.Rect(name_positions[i][0] - 50, name_positions[i][1] - 20, 100, 40)
                            if name_rect.collidepoint(mouse_pos):
                                selected_name_index = i  # Seleccionar nombre

                    # Verificar si una imagen fue seleccionada
                    for i, img in enumerate(loaded_images):
                        if not disabled[i]:  # Solo permitir seleccionar si no está deshabilitado
                            image_rect = pygame.Rect(image_positions[i][0], image_positions[i][1], 100, 100)
                            if image_rect.collidepoint(mouse_pos):
                                selected_image_index = i  # Seleccionar imagen

                    # Verificar si el nombre y la imagen coinciden
                    if selected_name_index is not None and selected_image_index is not None:
                        if selected_name_index == selected_image_index and selected_name_index not in matches:  # Emparejamiento correcto
                            matches.append(selected_name_index)  # Agregar el índice del emparejamiento
                            disabled[selected_name_index] = True  # Deshabilitar el par emparejado
                            disabled[selected_image_index] = True  # Deshabilitar el par emparejado
                        selected_name_index = None
                        selected_image_index = None

            # Actualizar pantalla
            pygame.display.flip()

    # Iniciar el juego
    game_loop()





```

```python
import pygame
import sys
import random

# Inicializa Pygame
pygame.init()

# Configuración de la pantalla
screen_width = 800
screen_height = 600
screen = pygame.display.set_mode((screen_width, screen_height))
pygame.display.set_caption('Novela Visual en Pygame')

# Colores
WHITE = (255, 255, 255)
BLACK = (0, 0, 0)
GRAY = (200, 200, 200)
LIGHT_BLUE = (173, 216, 230)
DARK_BLUE = (0, 102, 204)
RED = (255, 0, 0)  # Color rojo para líneas de error

# Carga de imágenes
background_image = pygame.image.load('imgs/Livingroom.jpg')
background_image = pygame.transform.scale(background_image, (screen_width, screen_height))

lola_image = pygame.image.load('imgs/Lola-removebg.png')
whiskers_image = pygame.image.load('imgs/Whiskers-removebg.png')
flashcard_image = pygame.image.load('imgs/1.png')
flashcard_image = pygame.transform.scale(flashcard_image, (screen_width-100, 400)) 


arrow_image = pygame.image.load('imgs/right-arrow.png')
arrow_image = pygame.transform.scale(arrow_image, (50, 50))  # Escala la flecha a un tamaño adecuado

# Fuentes para texto
font = pygame.font.Font(None, 40)

# Diálogos del juego
dialogues_english = [
    ("",""),
    ("Lola", "Hello! My name is Lola."),
    ("Lola", "I am a Poodle. I am happy!"),
    ("Lola", "This is my brother. His name is Whiskers."),
    ("Lola", "He is a cat. He is smart!"),
    ("Lola", "My grandma gave me a necklace."),
    ("Lola", "The collar is special. I love it!"),
    ("Lola", "Now, I go to sleep."),
    ("Lola", "Tomorrow is a new day!")
]

dialogues_spanish = [
    ("",""),
    ("Lola", "¡Hola! Me llamo Lola."),
    ("Lola", "Soy una Poodle. ¡Estoy feliz!"),
    ("Lola", "Este es mi hermano. Se llama Whiskers."),
    ("Lola", "¡Es un gato. Es inteligente!"),
    ("Lola", "Mi abuela me dio un collar."),
    ("Lola", "El collar es especial. ¡Lo amo!"),
    ("Lola", "Ahora, me voy a dormir."),
    ("Lola", "¡Mañana es un nuevo día!")
]

# Variables del juego
current_dialogue = 0
clock = pygame.time.Clock()
show_spanish = False  # Variable para controlar si mostrar en español

def draw_text(surface, text, font, color, rect, aa=False, bkg=None):
    """Función para dibujar texto que se ajusta automáticamente dentro de un rectángulo."""
    rect = pygame.Rect(rect)
    y = rect.top
    line_spacing = -2

    # Obtener altura de la fuente
    font_height = font.size("Tg")[1]

    while text:
        i = 1

        # Determinar si el texto puede caber en una línea
        while font.size(text[:i])[0] < rect.width and i < len(text):
            i += 1

        # Si hemos excedido el ancho, buscar el último espacio
        if i < len(text):
            i = text.rfind(" ", 0, i) + 1

        # Renderizar la línea
        if bkg:
            image = font.render(text[:i], aa, color, bkg)
            image.set_colorkey(bkg)
        else:
            image = font.render(text[:i], aa, color)

        surface.blit(image, (rect.left, y))
        y += font_height + line_spacing

        # Eliminar la línea de texto renderizada
        text = text[i:]

        # Si se salió del rectángulo, parar
        if y + font_height > rect.bottom:
            break

def draw_language_button():
    """Dibuja el botón de cambio de idioma en la esquina superior derecha."""
    button_rect = pygame.Rect(screen_width - 120, 20, 100, 40)  # Botón en la esquina superior derecha
    pygame.draw.rect(screen, LIGHT_BLUE, button_rect)  # Fondo del botón
    pygame.draw.rect(screen, DARK_BLUE, button_rect, 2)  # Borde del botón
    draw_text(screen, "Ayuda", font, BLACK, button_rect)  # Texto del botón
    return button_rect

def draw_back_button():
    """Dibuja el botón de retroceso en la esquina superior izquierda."""
    button_rect = pygame.Rect(20, 20, 100, 40)  # Botón en la esquina superior izquierda
    pygame.draw.rect(screen, LIGHT_BLUE, button_rect)  # Fondo del botón
    pygame.draw.rect(screen, DARK_BLUE, button_rect, 2)  # Borde del botón
    draw_text(screen, "Atrás", font, BLACK, button_rect)  # Texto del botón
    return button_rect

def draw_arrow_button():
    """Dibuja el botón de la flecha para cambiar al juego de emparejar parejas."""
    if current_dialogue == len(dialogues_english) - 1:  # Si estamos en el último diálogo
        arrow_rect = pygame.Rect(screen_width - 100, screen_height - 100, 50, 50)  # Posición de la flecha
        screen.blit(arrow_image, arrow_rect)  # Dibuja la imagen de la flecha
        return arrow_rect  # Devuelve el rectángulo para la detección de colisiones
    return None  # Si no es el último diálogo, no hay botón

def game_loop():
    global current_dialogue, show_spanish

    running = True
    language_button = draw_language_button()  # Dibuja el botón de idioma una vez
    back_button = draw_back_button()  # Dibuja el botón de retroceso una vez

    while running:
        for event in pygame.event.get():
            if event.type == pygame.QUIT:
                pygame.quit()
                sys.exit()

            # Detectar clics
            if event.type == pygame.MOUSEBUTTONDOWN:
                mouse_pos = pygame.mouse.get_pos()  # Obtener la posición del mouse
                if language_button.collidepoint(mouse_pos):
                    show_spanish = not show_spanish  # Alternar el idioma
                elif back_button.collidepoint(mouse_pos):  # Si se hace clic en el botón de retroceso
                    if current_dialogue > 0:  # Asegúrate de no retroceder más allá del primer diálogo
                        current_dialogue -= 1
                elif arrow_button and arrow_button.collidepoint(mouse_pos):  # Si se hace clic en la flecha
                    running = False  # Detener el loop de la novela visual
                    matching_game()  # Llama al juego de emparejar parejas
                    
                else:
                    # Avanza al siguiente diálogo si no es el botón
                    if current_dialogue < len(dialogues_english) - 1:
                        current_dialogue += 1

        # Dibuja el fondo
        screen.blit(background_image, (0, 0))

        # Dibuja a Lola y Whiskers
        if current_dialogue == 3 or current_dialogue == 4:  # Si es el diálogo donde presenta a Whiskers
            screen.blit(lola_image, (0, 75))  # Mueve a Lola a la izquierda
            screen.blit(whiskers_image, (350, 75))  # Aparece Whiskers a la derecha
        elif current_dialogue==0:
            screen.blit(flashcard_image, (60, 75))
        else:
            screen.blit(lola_image, (150, 75))  # Mantén a Lola en el centro en otros diálogos

        # Dibuja el cuadro de diálogo
        pygame.draw.rect(screen, BLACK, (50, 465, 720, 190))
        pygame.draw.rect(screen, WHITE, (60, 475, 700, 130))

        # Muestra el diálogo actual en el idioma correspondiente
        if current_dialogue < len(dialogues_english):
            speaker, text = dialogues_spanish[current_dialogue] if show_spanish else dialogues_english[current_dialogue]
            draw_text(screen, f"{speaker}:", font, BLACK, (80, 485, 700, 45))
            draw_text(screen, text, font, BLACK, (80, 520, 700, 130))

        # Dibuja el botón de cambio de idioma y retroceso
        language_button = draw_language_button()  # Dibuja el botón cada vez en el bucle
        back_button = draw_back_button()  # Dibuja el botón de retroceso cada vez en el bucle

        # Dibuja el botón de flecha si es el último diálogo
        arrow_button = draw_arrow_button()

        # Actualiza la pantalla
        pygame.display.flip()

        # Controla la velocidad del bucle
        clock.tick(60)

# Iniciar el juego
game_loop()




```
